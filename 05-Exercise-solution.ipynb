4. write a function to reformat a name to only contain "firstname lastname" (look for current format in the name column)

```python
#Fra Vincent på Python Discord kanalen:

def reformat_name(name):
    first_name = name.split(', ')[1].split(' ')[1].strip()
    last_name = name.split(',')[0].strip()
    
    return f'{first_name} {last_name}'

df['Name'] = df['Name'].apply(reformat_name)

print(df)
```

```python
def get_lastname(name):
    lastname = name.split(',')[0].strip()
    
    return f'{lastname}'
                          
df['Name'] = df['Name'].apply(get_lastname)

print(df)
                        
```
